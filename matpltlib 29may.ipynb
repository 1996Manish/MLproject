```python
#
import matplotlib.pyplot as plt
from matplotlib import style
style.use("ggplot")
x1=[2,3,4]
y1=[3,2,4]

x2=[1,2,3]
y2=[3,2,1]
plt.plot(x1,y1,linewidth=3,color='y',label='1st one')      # for thickness of line we use linewidth
plt.plot(x2,y2,linewidth=3,color='g',label='2nd one')
plt.title('my graph')                                      # giving title
plt.xlabel('x-axis')                                       # naming  x-axis
plt.ylabel('y-axis')                                       #naming y-axis
plt.legend()                                               # legend is  to show  the label like '1st one'
plt.grid(color='k')                                        # for changing the color of background  square grid
plt.show()
```

```python
#BAR
import matplotlib.pyplot as plt
plt.bar((1,2,3,4,4,1),(5,3,4,5,3,6),color='g',label='ages')
plt.bar((5,8,2,6,7,3),(2,1,9,3,4,8),color='y',label='number')
plt.xlabel('my x')
plt.ylabel('my y')
plt.title('my bar')
plt.axis([0,10,0,12])    # increasing the range of axis
plt.legend()
plt.show()
```

```python
#  histogram
import matplotlib.pyplot as plt
px=[5,51,42,2,78,44,15,14,4,16,8,45,12,78,98,78,4,1,3,42,45,32,4,52,42,234,2,3,2,2,42,2]
py=[0,20,40,60,80,100,120,140,160,180,200,220,240,260]
plt.hist(px,py,rwidth=0.8,color='b',label='my histogram')
plt.title('my histo')
plt.legend()
plt.show()
```

```python
#scatter plt
import matplotlib.pyplot as plt

x=[1,2,3,3,4,2,3]
y=[1,3,2,2,2,2,2]
plt.scatter(x,y,color='k',marker='v',s=100)     # marker is used to change the shape of dot
plt.title('scatter')
plt.show()
```

#stack/area plot


```python
import matplotlib.pyplot as plt

days=[1,2,3,4,5]
sleeping=[3,6,7,8,8]
eating=[1,6,2,1,3]
working=[2,4,1,13,3]
plt.plot([],[],color='m',label='sleeping')
plt.stackplot(days,sleeping,eating,working,color='y')
plt.title('areaplot')
plt.xlabel('x axis')
plt.xlabel('y axis')
plt.legend()
plt.show()
```

```python
#pie chart
slices=[7,2,2,3]
activities=['sleeping','eating','working','playing']
cols=['c','m','r','b']   
plt.pie(slices,labels=activities,colors=cols,shadow='true',startangle=30)
plt.title('my pie-chart')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)
t1=np.arange(0,0.5,2)
t2=np.arange(0,0.6,3)
# where nrows and ncols are used to notionally split the figure into nrows& ncols  sub axes and plot number is used to  idenify the particular
#subplot that this function is to create within the notaional grid plot_number starts at l,increments rows  
plt.subplot()
plt.plot(t1,f(t1),'bo',t2,f(t2))
plt.subplot(212)
plt.plot(t2,np.cos(2*np.pi*t2))
plt.show()
```
